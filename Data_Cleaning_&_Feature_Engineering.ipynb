```python
stats = pd.read_csv("NFL_STATS.csv")
```

```python
team_summary = (
    stats.groupby(["season", "team"])
    .agg(
        games=("week", "count"),
        avg_points_for=("score_for", "mean"),
        avg_points_against=("score_against", "mean"),
        avg_yards=("yards", "mean"),
        avg_pass_yards=("pass_yards", "mean"),
        avg_rush_yards=("rush_yards", "mean"),
        avg_fumbles=("fumbles", "mean"),
        avg_interceptions=("interceptions", "mean"),
        third_down_comp=("third_down_comp", "sum"),
        third_down_att=("third_down_att", "sum"),
        fourth_down_comp=("fourth_down_comp", "sum"),
        fourth_down_att=("fourth_down_att", "sum"),
        redzone_comp=("redzone_comp", "sum"),
        redzone_att=("redzone_att", "sum"),
    )
    .reset_index()
)

team_summary["point_diff"] = team_summary["avg_points_for"] - team_summary["avg_points_against"]
team_summary["turnovers"] = team_summary["avg_fumbles"] + team_summary["avg_interceptions"]
team_summary["third_down_pct"] = (team_summary["third_down_comp"] / team_summary["third_down_att"]) * 100
team_summary["fourth_down_pct"] = (team_summary["fourth_down_comp"] / team_summary["fourth_down_att"]) * 100
team_summary["redzone_pct"] = (team_summary["redzone_comp"] / team_summary["redzone_att"]) * 100
team_summary = team_summary.drop(columns=[
    "avg_fumbles","avg_interceptions",
    "third_down_comp","third_down_att",
    "fourth_down_comp","fourth_down_att",
    "redzone_comp","redzone_att"
])
team_summary = team_summary.round(1)
```

```python
combined_summary = (
    stats.groupby("team")
    .agg(
        games=("week", "count"),
        avg_points_for=("score_for", "mean"),
        avg_points_against=("score_against", "mean"),
        avg_yards=("yards", "mean"),
        avg_pass_yards=("pass_yards", "mean"),
        avg_rush_yards=("rush_yards", "mean"),
        avg_fumbles=("fumbles", "mean"),
        avg_interceptions=("interceptions", "mean"),
        third_down_comp=("third_down_comp", "sum"),
        third_down_att=("third_down_att", "sum"),
        fourth_down_comp=("fourth_down_comp", "sum"),
        fourth_down_att=("fourth_down_att", "sum"),
        redzone_comp=("redzone_comp", "sum"),
        redzone_att=("redzone_att", "sum"),
    )
    .reset_index()
)

combined_summary["point_diff"] = combined_summary["avg_points_for"] - combined_summary["avg_points_against"]
combined_summary["turnovers"] = combined_summary["avg_fumbles"] + combined_summary["avg_interceptions"]
combined_summary["third_down_pct"] = (combined_summary["third_down_comp"] / combined_summary["third_down_att"]) * 100
combined_summary["fourth_down_pct"] = (combined_summary["fourth_down_comp"] / combined_summary["fourth_down_att"]) * 100
combined_summary["redzone_pct"] = (combined_summary["redzone_comp"] / combined_summary["redzone_att"]) * 100
combined_summary = combined_summary.drop(columns=[
    "avg_fumbles","avg_interceptions",
    "third_down_comp","third_down_att",
    "fourth_down_comp","fourth_down_att",
    "redzone_comp","redzone_att"
])
combined_summary = combined_summary.round(1)
```

```python
import pandas as pd
import numpy as np

INPUT_CSV = "NFL_STATS_22-24_swapped.csv"
OUTPUT_CSV = "nfl_2022_team_summary.csv"
df = pd.read_csv(INPUT_CSV)
df = df[df['season'].astype(str) == '2022'].copy()

num_cols = [
    'score_for', 'score_against', 'yards', 'pass_yards', 'rush_yards',
    'third_down_comp', 'third_down_att',
    'fourth_down_comp', 'fourth_down_att',
    'redzone_comp', 'redzone_att',
    'fumbles_lost', 'fumbles_forced', 'interceptions_thrown', 'interceptions_forced'
]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce').fillna(0)

grp = df.groupby('team')

summary = pd.DataFrame({
    'team': grp.size().index,
    'games': grp.size().values,
    'avg_points_for': grp['score_for'].mean().values,
    'avg_points_against': grp['score_against'].mean().values,
    'avg_yards': grp['yards'].mean().values,
    'avg_pass_yards': grp['pass_yards'].mean().values,
    'avg_rush_yards': grp['rush_yards'].mean().values,
    'turnovers': (grp['fumbles_lost'].sum() + grp['interceptions_thrown'].sum()).values,
    'takeaways': (grp['fumbles_forced'].sum() + grp['interceptions_forced'].sum()).values,
    'point_diff': (grp['score_for'].sum() - grp['score_against'].sum()).values,
})

summary['turnover_diff'] = summary['takeaways'] - summary['turnovers']
third_comp = grp['third_down_comp'].sum()
third_att = grp['third_down_att'].sum()
summary['third_down_pct'] = np.where(third_att > 0, (third_comp / third_att) * 100.0, 0.0)
fourth_comp = grp['fourth_down_comp'].sum()
fourth_att = grp['fourth_down_att'].sum()
summary['fourth_down_pct'] = np.where(fourth_att > 0, (fourth_comp / fourth_att) * 100.0, 0.0)
rz_comp = grp['redzone_comp'].sum()
rz_att = grp['redzone_att'].sum()
summary['redzone_pct'] = np.where(rz_att > 0, (rz_comp / rz_att) * 100.0, 0.0)
numeric_cols = summary.columns.drop('team')
summary[numeric_cols] = summary[numeric_cols].round(1)
summary = summary.sort_values('team').reset_index(drop=True)
summary.to_csv(OUTPUT_CSV, index=False)
print("Wrote:", OUTPUT_CSV)
print(summary.head(32).to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("updated_all.csv")
df_sorted = df.sort_values('win_numeric', ascending=False)

top_5 = df_sorted.head(5)
plt.figure(figsize=(8,5))
plt.bar(top_5['team'], top_5['win_numeric'], color='green')
plt.title('Top 5 Teams by Wins (2022-2024)')
plt.ylabel('Total Wins')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, df['win_numeric'].max() + 2)
plt.show()

bottom_5 = df_sorted.tail(5)
plt.figure(figsize=(8,5))
plt.bar(bottom_5['team'], bottom_5['win_numeric'], color='red')
plt.title('Bottom 5 Teams by Wins (2022-2024)')
plt.ylabel('Total Wins')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, df['win_numeric'].max() + 2)
plt.show()
```
